```python
# 데이터 로딩
import numpy as np
import pandas as pd
import torch
import torch.nn
```

```python
# 데이터 로딩 => 데이터프레임
url = 'http://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'
column_names = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight',
                'Acceleration', 'Model Year', 'Origin']


df = pd.read_csv(url, names=column_names,
                 na_values = "?", comment='\t',
                 sep=" ", skipinitialspace=True)


df.tail()
```

```python
# Nan값처리
# 확인
# df.head()
df.isna().sum()
df[df['Horsepower'].isna()]
```

```python
# df=df.dropna(axis=0)
df=df.dropna()
df[df['Horsepower'].isna()]
```

```python
#  인덱스 연속값으로 재설정 
df=df.reset_index(drop=True)
```

```python
# 학습 / 테스트 분리 
from sklearn.model_selection import train_test_split

df_train, df_test=train_test_split(
    df,
    train_size=0.8,
    random_state=1
)
```

```python
# 학습데이터 기본 통계량 확인 
train_stats=df_train.describe().transpose() #mean 값 가지고 오기 
train_stats
```

```python
# 연속값을 가지고 있는 피처의 정규화
# cylinder => iloc 적용 
numeric_column_names = ['Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration']


df_train_norm, df_test_norm = df_train.copy(), df_test.copy()


for col_name in numeric_column_names:
    mean = train_stats.loc[col_name, 'mean']
    std  = train_stats.loc[col_name, 'std']
    df_train_norm.loc[:, col_name] = (df_train_norm.loc[:, col_name] - mean)/std
    df_test_norm.loc[:, col_name] = (df_test_norm.loc[:, col_name] - mean)/std

```

```python
# df_train_norm['Origin'].unique() # array([1, 2, 3], dtype=int64)
# 토치에도 원핫 처리
from torch.nn.functional import one_hot

total_origin = len(set(df_train_norm['Origin']))


origin_encoded = one_hot(torch.from_numpy(df_train_norm['Origin'].values) % total_origin)
x_train_numeric = torch.tensor(df_train_norm[numeric_column_names].values)
x_train = torch.cat([x_train_numeric, origin_encoded], 1).float()


origin_encoded = one_hot(torch.from_numpy(df_test_norm['Origin'].values) % total_origin)
x_test_numeric = torch.tensor(df_test_norm[numeric_column_names].values)
x_test = torch.cat([x_test_numeric, origin_encoded], 1).float()
```

```python
# 데이터, 레이블 추출
y_train = torch.tensor(df_train_norm['MPG'].values).float()
y_test = torch.tensor(df_test_norm['MPG'].values).float()
```

```python
# 데이터셋 -> 데이터로더 생성 : 신경망 데이터 준비
from torch.utils.data import TensorDataset, DataLoader

train_ds=TensorDataset(x_train, y_train) # 학습데이터셋 생성
batch_size=8
torch.manual_seed(1)
train_dl=DataLoader(train_ds, batch_size, shuffle=True)
```

```python
# 신경망 구성 : forward까지 처리
# 히든층 구조 설명
# input 개수 설정
hidden_units=[8,4,2] # 2차원 : batch_size=8 -> 예측값은 batch_size에 의하여 결정
input_size=x_train.shape[1] # torch.Size([313, 9]) -> 9

all_layers=[]

for hidden_unit in hidden_units:
    layer=nn.Linear(input_size, hidden_unit) # (9,8)
    all_layers.append(layer)
    all_layers.append(nn.ReLU())
    input_size=hidden_unit # hidden_unit=8 -> input_size에 대입
    
all_layers.append(nn.Linear(hidden_units[-1],1))

model=nn.Sequential(*all_layers)
model
```

```python
# 오차함수 정의, 옵티마이저 정의
loss_fn=nn.MSELoss()
optimizer=torch.optim.SGD(model.parameters(), lr=0.001) # lr : 학습율, optimizer : 오차 최소화
```

```python
# 학습 : 반복
num_epochs=200
log_epochs=10 # log : 처리 결과를 화면에 출력

for epoch in range(num_epochs): # 200번 반복
    loss_hist_train=0
    for X_batch, y_batch in train_dl:
        # forward -> pred
        pred=model(X_batch)[:,0]
        loss=loss_fn(pred, y_batch)
        loss.backward() # 미분
        optimizer.step() # 수정
        optimizer.zero_grad() # weight 초기화
        loss_hist_train += loss.item() # 오차 추출 후 누적합
    if epoch % log_epochs==0:
        print(f'에포크 : {epoch}, 손실 : {loss_hist_train/len(train_dl):.4f}')
```
