```python
# cctv 현황데이터 읽기
import numpy as np 
import pandas as pd

# 데이터 로딩
CCTV_Seoul = pd.read_csv(
    './data/data/01. CCTV_in_Seoul.csv'
    , encoding='utf-8'
)
CCTV_Seoul.head(3)
```

```python
# 정렬해서 데이터 확인 : 소게 ASC
CCTV_Seoul.sort_values(
    by= '소계' #컬럼명 
    , ascending=True #default
).head()
```

```python
# DESC 
CCTV_Seoul.sort_values(
    by='소계'
    , ascending=False
).head()
```

```python
# 최근 증가울 구하고 컬럼추가 
CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2014년'] \
                          + CCTV_Seoul['2015년'] \
                          + CCTV_Seoul['2016년']) \
                            / CCTV_Seoul['2013년도 이전'] * 100           
CCTV_Seoul.head()
                        
```

```python
# 최근 3년간 많이 증가한 구를 확인 : 정렬
CCTV_Seoul.sort_values(
    by='최근증가율'
    , ascending=False
).head()
```

```python
# 컬럼명 변경 : 기관명=> 구별 변경
CCTV_Seoul.rename(
    columns={
        CCTV_Seoul.columns[0] : '구별'
    }
    , inplace=True
)
CCTV_Seoul.head()
```

```python
# 인구현황
POP_Seoul = pd.read_excel(
    './data/data/01. population_in_Seoul.xls'
    , header=2
    , usecols='B,D,G,J,N'
)
POP_Seoul.head()
```

```python
# 컬럼명 변경 : RENAME()
POP_Seoul.rename(
      columns={
        POP_Seoul.columns[0] : '구별'
        , POP_Seoul.columns[1] : '인구수'
        , POP_Seoul.columns[2] : '한국인'
        , POP_Seoul.columns[3] : '외국인'
        , POP_Seoul.columns[4] : '고령자'
    }
    , inplace=True # 원본 반영영
)
POP_Seoul.head()
  
```

```python
# 데이터프레임에서 행, 컬럼삭제 : Drop([정수인덱스], 축 axis=0,1 , inplace)
# axis =0 : 행방향 
# axis =1 : 열방향 
# 합계라는 행 삭제 
POP_Seoul.drop(
    [0] #[0] : 0번행 삭제, 리스트 : 여러개 삭제 가능, 하나로 묶는다
    , axis=0
    # , #inplace=True
).head()
# 숫자도 named 가 된다 
```

```python
# 데이터프레임에서 행, 컬럼삭제 : Drop([정수인덱스], 축 axis=0,1 , inplace)
# axis =0 : 행방향 
# axis =1 : 열방향 
# 합계라는 행 삭제 
POP_Seoul.drop(
    [0] #[0] : 0번행 삭제, 리스트 : 여러개 삭제 가능, 하나로 묶는다
    , axis=0
    , inplace=True
) # return None -> .head() error
# 숫자도 named 가 된다 
```

```python
# inplace 사용시 주의
# drop(inplace=False) : 복사된 오브젝트에서 삭제된 데이터 프레임 반환, head()가능
# drop(inplace=True) : 원본에 반영이 된다 , 복사 없다, None 반환 
```
