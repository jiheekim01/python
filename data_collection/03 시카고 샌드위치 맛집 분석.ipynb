```python
# 샌드위치 메인 페이지 데이터 추출
import requests

# 메인 페이지 URL 

url_base='https://www.chicagomag.com'
# 현재는 샌드위치 메인 페이지 URL
url_sub='/Chicago-Magazine/November-2012/Best-Sandwiches-Chicago'

url=url_base+url_sub

# header 
headers={
    'User-Agent' : 'Mozilla/5.0'
}

response = requests.get(
    url=url
    , headers=headers
).text

# 파싱
soup = BeautifulSoup(response,'html.parser')

# 데이터 추출
# 순위, 메인메뉴, 카페명, 세부페이지 이동 URL 
ranks=[]
main_menus=[]
cafe_names=[]
url_address=[] # 상세로 들어가는 URL

# 50개 정보가 들어 있는 리스트 
list_soup = soup.find_all('div', class_='sammy')

# 반복 처리해서 데이터 추출하고 각 리스트에 저장
for item in list_soup: # 50번 반복
    # 순위 추출해서 저장
    ranks.append(item.find(class_='sammyRank').text)
    # 메인메뉴와 카페명 추출해서 저장
    # 'BLT\nOld Oak Tap\nRead more '
    tmp_str=item.find(class_='sammyListing').text

    main_menus.append(tmp_str.split('\n')[0]) # BLT
    cafe_names.append(tmp_str.split('\n')[1]) # Old Oak Tap
    
    tmp_url=item.find('a')['href'] # sub의 URL 추출
    # https:// 시작하거나 /Chicago 시작할 수 있다
    if tmp_url.startswith('https://'): # 전체 URL이 다 있는 경우
        url_address.append(tmp_url)
    else: # https://로 시작하지 않으면
        url_address.append(url_base+tmp_url)


```

```python
# 데이터 수집 확인
print(ranks[:5])
print(main_menus[:5])
print(cafe_names[:5])
print(url_address[:5])
```

```python
url_address
```

```python
# DataFrame 생성
data_df=pd.DataFrame({
    'Rank':ranks,
    'Menu':main_menus,
    'Cafe':cafe_names,
    'Url':url_address
})
data_df.head()
```

```python
# data_df 파이에 저장
data_df.to_csv(
    './data/03. 시카고 샌드위치 메인 페이지 데이터.csv'
    , encoding='utf-8'
)
```

```python
import requests
from bs4 import BeautifulSoup

url_str=data_df['Url'][0]

headers={
    'User-Agent':'Mozilla/5.0'
}

response=requests.get(url=url_str, headers=headers).text

# text => DOM 파싱
tmp_soup=BeautifulSoup(response,'html.parser')

# 가격, 주소가 있는 영역 추출 <p class='addy'>가격, 주소 </p>
print_address=tmp_soup.find('p', class_='addy').text
print_address
```

```python
# ['2109', 'W.', 'Chicago', 'Ave.,'] 리스트의 값들을
# 하나의 문자열로 연결하는 함수 : join()
' '.join(print_address.split()[1:-2])
```

```python
# url이 50개다. 50반복 처리
# 반복시 반복상태를 화면에 표시 : tqdm
from tqdm import tqdm

# 가격, 주소 저장
price=[]
address=[]

for idx in tqdm(data_df['Url'].index): # [0,1,,,,,,,,48,49]
    url=data_df['Url'][idx] # url 한개씩 추출
    
    headers={
        'User-Agent':'Mozilla/5.0'
    }
    response=requests.get(url=url,headers=headers).text
    soup=BeautifulSoup(response,'html.parser')
    get_str=soup.find('p', class_='addy').text
    price.append(get_str.split()[0][:-1]) # $10
    address.append(' '.join(get_str.split()[1:-2])) # '2109 W. Chicago Ave.,'
```

```python
price[:5], address[:5]
```

```python
# data_df DataFrame 위 데이터 추가
data_df['Price']=price
data_df['Address']=address

data_df.head()
```

```python
# Rank	Menu	Cafe	Price	Address
# data_result_df=
data_result_df=data_df[['Rank','Menu','Cafe','Price','Address']]
data_result_df.head()
```

```python
# Rank column => index 
data_result_df.set_index(
    'Rank' # 인덱스로 보낼 컬럼들 지정
    , inplace=True # 원본 데이터프레임에 반영
)
data_result_df.head()
```

```python
# lat, lng 구한다
# 주소에 Multiple => 주소가 없다
lat=[]
lng=[]

for idx in tqdm(data_result_df.index): # 50번 반복
    if data_result_df['Address'][idx] != 'Multiple':
        # 주소가 있는 경우 : 구글로 보내서 lat, lng 구한다
        target_name=data_result_df.loc[idx, 'Address'] + 'Chicago'
        gmaps_output=gmaps.geocode(target_name) # 주소가 들어있는 리스트
        location_output=gmaps_output[0].get('geometry') # 딕셔너리 반환
        lat.append(location_output['location']['lat'])
        lng.append(location_output['location']['lng'])
    else:
        # 주소가 없는 경우 : lat, lng => NaN처리
        lat.append(np.nan)
        lng.append(np.nan)
```

```python
len(lat), len(lng)
```

```python
data_result_df['Lat']=lat
data_result_df['Lng']=lng
data_result_df.head()
```

```python
# 지도 표시
# 지도 생성
map=folium.Map(
    location=[ # 지도 중심 설정
        data_result_df['Lat'].mean()
        , data_result_df['Lng'].mean()
    ]
    , zoom_start=11
)

# 50개 맛집 마커 출력
for idx in data_result_df.index:
    if data_result_df['Address'][idx] != 'Multiple':
        folium.Marker(
            [data_result_df['Lat'][idx], data_result_df['Lng'][idx]]
            , popup=data_result_df['Cafe'][idx]
        ).add_to(map)

map
```
