```python
# # 2. ê¸°ë³¸ì ì¸ get ìš”ì²­ 
# # get(url) : í•´ë‹¹ urlì— http getìš”ì²­ì„ ë³´ë‚¸ë‹¤
# # status code : ì‘ë‹µ ì½”ë“œ (200(ì •ìƒì²˜ë¦¬),400(ìš”ì²­ì˜ëª»),500(ì„œë²„ì—ëŸ¬))
# # text : ì›¹í˜ì´ì§€ë¥¼ ë¬¸ìì—´ë¡œ ë°˜í™˜, 
import requests

# # í¬ë¡¤ë§ í•œ ì›¹í˜ì´ì§€ URL ì§€ì • 
url = 'https://www.naver.com' #ë©”ì¸ í˜ì´ì§€

# # get ìš”ì²­
response = requests.get(url)

# # ì‘ë‹µ ì½”ë“œ í™•ì¸
print('ì‘ë‹µ ìƒíƒœ ì½”ë“œ : ',response.status_code)

# # í˜ì´ì§€ ë‚´ìš© ì¶œë ¥(html ì½”ë“œ)
print('='*100)
print(response.text) #ë³€ìˆ˜: attribute
print('='*100)
print('ë‹¨ì–´ìˆ˜ : ',len(response.text))
```

```python
# ìš”ì²­ í—¤ë”(user Agent) ë³€ê²½í•˜ê¸°
# ì‚¬ìœ : ì¼ë¶€ ì›¹ì‚¬ì´íŠ¸ëŠ” user-Agent ê°€ ì—†ëŠ” ìš”ì²­ì„ ì°¨ë‹¨í•  ìˆ˜ ìˆë‹¤
# ì½”ë“œ ì‹¤í–‰ ì•ˆëœë‹¤
# ì¼ë¶€ ì‚¬ì´íŠ¸ëŠ” ê¸°ë³¸ì ìœ¼ë¡œ ë´‡ ìš”ì²­ì„ ì°¨ë‹¨í•œë‹¤.
# ì„œë²„ ì…ì¥ì—ì„œ ì½”ë“œë¡œ ì‹¤í–‰í•˜ëŠ” ê²ƒì´ ì•„ë‹ˆê³ , ë¸Œë¼ìš°ì €ì—ì„œ ì ‘ì†í•œ ê²ƒì²˜ëŸ¼ ë³´ì´ë„ë¡ ì„¤ì •
  
import requests

url = 'http://naver.com'

# ìš”ì²­ í—¤ë” ì„¤ì •(ì¼ë°˜ì ì¸ ë¸Œììš°ì € ì‚¬ìš©í•˜ëŠ” ê²ƒì²˜ëŸ¼ ì„¤ì • )
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36" }

# get ìš”ì²­ 
response = requests.get(url, headers=headers)

# respose ìƒíƒœ í™•ì¸, íƒìŠ¤íŠ¸ ì¶”ì¶œ
print('ì‘ë‹µ ìƒíƒœ ì½”ë“œ : ', response.status_code)
print(response.text[:100])
```
