#### requests 모듈

1. 모듈 소개 
    1. requests 모듈을 http 요청을 쉽게 보내고 결과를 응답받는 기능 
        1. 요청 (request) : get, post 
        2. 응답 (response) 

```python
# # 2. 기본적인 get 요청 
# # get(url) : 해당 url에 http get요청을 보낸다
# # status code : 응답 코드 (200(정상처리),400(요청잘못),500(서버에러))
# # text : 웹페이지를 문자열로 반환, 
import requests

# # 크롤링 한 웹페이지 URL 지정 
url = 'https://www.naver.com' #메인 페이지

# # get 요청
response = requests.get(url)

# # 응답 코드 확인
print('응답 상태 코드 : ',response.status_code)

# # 페이지 내용 출력(html 코드)
print('='*100)
print(response.text) #변수: attribute
print('='*100)
print('단어수 : ',len(response.text))
```

```python
# 요청 헤더(user Agent) 변경하기
# 사유: 일부 웹사이트는 user-Agent 가 없는 요청을 차단할 수 있다
# 코드 실행 안된다
# 일부 사이트는 기본적으로 봇 요청을 차단한다.
# 서버 입장에서 코드로 실행하는 것이 아니고, 브라우저에서 접속한 것처럼 보이도록 설정
  
import requests

url = 'http://naver.com'

# 요청 헤더 설정(일반적인 브자우저 사용하는 것처럼 설정 )
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/122.0.0.0 Safari/537.36" }

# get 요청 
response = requests.get(url, headers=headers)

# respose 상태 확인, 택스트 추출
print('응답 상태 코드 : ', response.status_code)
print(response.text[:100])
```

```python
# 구글 검색을 파이선코드로 
# https://www.google.com/search?q=%EB%94%A5%EB%9F%AC%EB%8B%9D&oq=%EB%94%A5%EB%9F%AC%EB%8B%9D&gs_lcrp=EgZjaHJvbWUyDAgAEEUYORixAxiABDIHCAEQABiABDIHCAIQABiABDIKCAMQABiLAxiABDIKCAQQABiLAxiABDIHCAUQABiABDIHCAYQLhiABDIHCAcQABiABDIHCAgQABiABDIHCAkQABiABNIBCjkwMDUzajFqMTWoAgiwAgHxBW0saVKl1T1u8QVtLGlSpdU9bg&sourceid=chrome&ie=UTF-8
query = '딥러닝'
url = 'https://www.google.com/search'

# get 요청에 사용할 파라미터 설정 : 딕셔너리 형태설정
params = {'q':'딥러닝'}

# 요청보내기 : get 
response = requests.get(url, params=params)

print(response.text)
```

```python
# 5. json 데이터 요청 및 처리 
# 증권사 api, REST Api -> 응답이 데이터가 온다 (json)

# 게시판 1번 요청 => jason 응답
url = 'https://jsonplaceholder.typicode.com/posts/1'

response = requests.get(url)

# 여기까지 response는 결과가 json 형태이긴 하지만 문자열 
# 위 문자열을 딕셔너리로 변환
data = response.json()

print('제목 : ', data['title'])
print('내용 : ', data['body'])
```

```python
# 크롤링시 예외처리 및 오류 대응 방법 
# 이 코드는 실행 안됨 
url = 'http://example.com'

# 예외 처리 
try:
    response = requests.get(url, timeout=5) # 5초 동안 응답을 받겟다
    response.raise_for_status() # 4xx, 5xx 응답시 예외 발생
    print(response.text[:100])
except requests.Timeout: #타임아웃 되엇을때 처리
    print('요청 시간 초과') 
except requests.HTTPError as e: # 네트웍 오류
    print('HTTP 오류 : ',e)
except requests.RequestException as e:
    print('요청 중 오류 발생 : ', e) # 요청 오류
    
```
