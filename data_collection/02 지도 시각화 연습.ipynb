#### 서울시 범죄 현황 지도 시각화

```python
# 서울시 행정구역 데이터, 범죄 현환 데이터
import numpy as np
import pandas as pd
import folium
import json
import pickle

geo_path = './data/data/02. skorea_municipalities_geo_simple.json'
geo_str = json.load(
    open(geo_path, encoding='utf-8')
)

crime_anal_norm = pd.read_csv(
    './data/data/02. crime_in_Seoul_final.csv'
    , encoding='utf-8'
    , index_col=0
) 

crime_anal_norm.head()
```

```python
print(type(geo_str))
print(geo_str['features'])
print(geo_str['features'][0])
print(geo_str['features'][0]['id'])
```

```python
# 라이브러리 버전 및 설치 경로 확인 
print(folium.__version__)
print(folium.__path__)

print(pd.__version__)
```

```python
seoul_map = folium.Map(
    location=[37.5502, 126.982]  # 지도 중심
    , zoom_start=11
)
# seoul_map

folium.Choropleth(
    geo_data=geo_str
    , data=crime_anal_norm['범죄']
    , columns=[crime_anal_norm.index, crime_anal_norm['범죄']]
    , key_on='feature.id'  #구이름
).add_to(seoul_map)

seoul_map
```

```python
# 경찰서별 검거현황과 구별 범죄 발생 현황 지도에 표시
# 경찰서의 위치 정보 
'./data/data/02. crime_in_Seoul_include_gu_name.csv'

crime_anal_raw = pd.read_csv(
    './data/data/02. crime_in_Seoul_include_gu_name.csv'
    , encoding='utf-8'
    , index_col=0
    
)

# 경찰서 위치 정보 : station_lat, station.lng

```

```python
with open('./data/data/station_lat.pickle', 'rb') as f:
          station_lat = pickle.load(f)

with open('./data/data/station_long.pickle', 'rb') as f:
    station_lng = pickle.load(f)
print(station_lat)
print(station_lng)
```

```python
# crime_anal_raw dataframe append
crime_anal_raw['lat'] = station_lat
crime_anal_raw['lng'] = station_lng

crime_anal_raw.head()
```

```python
# 검거 컬럼을 추출 리스트
# cols = ['살인 검거','강도 검거','강간 검거','절도 검거','폭력 검거']
cols = [col for col in crime_anal_raw.columns if '검거' in col]
tmp_df = crime_anal_raw[cols] / crime_anal_raw[cols].max()
tmp_df
```

```python
print(crime_anal_raw[cols].head())
print(crime_anal_raw[cols].max())
```

```python
# 지도에 경찰서 위치 표시
map = folium.Map(
    location=[37.5502, 126.982]
    , zoom_start=11
)

for idx in crime_anal_raw.index:
    folium.Marker(
        location=[crime_anal_raw['lat'][idx], crime_anal_raw['lng'][idx]]
    ).add_to(map)
    
map
```

```python
# 전체지돗현황생성성
map = folium.Map(
    location=[37.5502, 126.982]
    , zoom_start=11
)
# 행정경계 생성
folium.Choropleth(
    geo_data=geo_str  #서울시 구별 경계 데이터 
    , data= crime_anal_norm['범죄'] #범죄 컬럼 : 범죄 발생
    , columns=[crime_anal_norm.index, crime_anal_norm['범죄']]
    , key_on='feature.id'
).add_to(map)

for idx in crime_anal_raw.index:
    folium.CircleMarker(
        location=[crime_anal_raw['lat'][idx], crime_anal_raw['lng'][idx]]
        , radius=crime_anal_raw['검거'][idx]*10
        , color='#FF0000'
        , fill_color='#FF0000'
    ).add_to(map)
    
map
```
